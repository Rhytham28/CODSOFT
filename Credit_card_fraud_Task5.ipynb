As we can see that our data is completely imbalanced and if we do modelling, our model might get biased and can give the proper accuracy but with this imbalanced data, so we can face problem in classification.

So first we will try to balance the data by using undersampling technique

```python
#Splitting our dataset into dependent(X) and independent(Y) features
X = df.drop('Class', axis = 1)
X.head()
X.shape

y = df['Class']
y.head()
y.shape
```

```python
!pip install -U imbalanced-learn
```

```python
#Using undersampling methods
from imblearn import under_sampling
from imblearn.under_sampling import NearMiss
```

```python
#Implementing undersampling to handle imbalanced dataset
nm = NearMiss()
X_1, y_1 = nm.fit_resample(X, y)
X_1.shape
y_1.shape
```

```python
X_1.head()
y_1.head()
```

```python
plt.figure(figsize = (8, 6))
sns.countplot(x = y_1, hue = y_1, data = df)
```

We can see above that our data is equally balanced so we can apply any classification alogrithm

We will apply logistic regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_1, y_1, test_size = 0.2, random_state = 10)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model.score(X_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print("confusion_matrix:")
print(cm)
print("accuracy_score:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
